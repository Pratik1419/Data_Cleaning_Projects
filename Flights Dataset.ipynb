```python
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
import seaborn as sns
```

```python
flight_data = sns.load_dataset("flights")
flight_data
```

```python
flight_data.info()
```

```python
flight_data.head()
```

```python
# CHECKING IF WE HAVE NULL VALUES ON THE DATAFRAME...

flight_data.isnull().sum()
```

```python
# CHECKING IF WE HAVE ANY DUPLICATED VALUE ...

flight_data.duplicated().sum()
```

```python
flight_data.describe()
```

```python
# Count of year of flight...
counts_year= flight_data["year"].value_counts()
counts_year
```

```python
# FIND HOW MANY PASSENGERS HAVE USED FLIGHT IN A MONTH

grouped_month = flight_data.groupby('month')['passengers'].sum().head()
print(grouped_month)
```

```python
# FIND THE MEAN VALUE OF THE PASSENGERS ARE USING FLIGHT IN A YEAR

grouped_year = flight_data.groupby('year')['passengers'].mean().round(2)
grouped_year
```

```python
# VISUALIZING OUR DATA OF MONTH OVER PASSENGERS ON THE FLIGHT...

sns.set(style = 'whitegrid')
sns.boxplot(hue='month',y='passengers', data = flight_data, palette='Paired', legend=False)
```

```python
# VISUALIZING OUR DATA OF YEAR OVER PASSENGERS ON THE FLIGHT...
sns.set(style = 'whitegrid')
sns.boxplot(hue='year',y='passengers', data = flight_data, palette='Paired', legend=False)
plt.xlabel('Year')
```

```python
# CREATING A PIVOT_TABLE FOR PASSENGER WITH THE INDEX OF MONTH AND COLUMNS
monthly_passengers = flight_data.pivot_table(values='passengers', index='month', columns='year')
sns.heatmap(monthly_passengers)
```

```python
# Creating  lineplot using both year and passengers

sns.lineplot(x='year', y='passengers', data=flight_data)
```

```python
# Creating  lineplot using both month and passengers...
sns.lineplot(x='month', y = 'passengers', data = flight_data)
```

```python
# Creating violinplot using both month and passengers...

sns.violinplot(x='month', hue='month', y='passengers', data = flight_data, palette='autumn_r', legend=False)
```

```python
# # Creating lineplot using both month and passengers with grouping by year and months to find the mean ...

sns.lineplot(x='month', y='passengers', hue='year', data=flight_data.groupby(['year', 'month']).mean())
```

```python
# Creating hisplot of passengers...

sns.histplot(data= flight_data,hue='year',x='passengers', bins=10, palette='PuBuGn_r', legend=False)
```
