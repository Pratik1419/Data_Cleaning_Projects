```python
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
import seaborn as sns
```

```python
df = pd.read_csv("Books_Data_Clean.csv", encoding="latin1")
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"] > 1900]
```

```python
df.isna().sum()
```

```python
# DROP NULL VALUES FROM THE DATA SET.

df.dropna(subset= "Book Name", inplace= True)
```

```python
# Check if Duplicated is present on the dataset.
df.duplicated().sum()
```

```python
# Check unique values.
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of books in Each Genre")
plt.show()
```

```python
# Group by Author and calculate the average rating for each Author

df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
# Creating a boxplot of book ratings counts
sns.boxplot(x="genre", y="Book_ratings_count", data = df)
plt.xlabel("Genre")
plt.ylabel("Book Rating Count")
plt.title("Box Plot of book rating Count for Each Genre")
plt.show()
```

```python
# Creat a scatter plot of Sale price and unit sold
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price VS Units Sold")
plt.show()



# From this Scatter Plt we can see that when the sale price is low the sold unit is also low.
```

```python
language_counts = df["language_code"].value_counts()
language_counts
```

```python
plt.pie(language_counts, labels=language_counts.index, startangle=90, autopct="%1.1f%%")
plt.title("Language Distribution of books")
plt.show()
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False)
```

```python
# Now we are going to group by the language_code and count the number of books in each lanuage...
df.groupby("language_code").size().sort_values(ascending=False)
```

```python
# Now we are going to group by the author rating and find the average book rating count for each author rating

df.groupby("Author_Rating")["Book_ratings_count"].min()
```

```python
# Showing the scatter plt of Book_ratings_count and Book_ratings_rating from our dataframe
plt.scatter(df["Book_average_rating"], df["Book_ratings_count"])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Ratings Count")
plt.title("Scatter Plot of Book Average Rating VS Book Ratings Count")
plt.show()
```

```python
# bar table for total growther sale for each author...

total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum()
total_gross_sales_by_author.sort_values(ascending=False).head(9).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales for Each Author")
plt.show()
```

```python
sns.boxplot(x="Author_Rating", y="units sold", data = df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Box Plot of units Sold For Each Author Rating")
plt.show()
```

```python
# Showing the Publishing Year and how many unit are sold on the each punlishing years from the dataframe

df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker = "o")
plt.xlabel("Publishing Year")
plt.ylabel("Total Units Sold")
plt.title("Total Units Sold Over The Years")
plt.show()
```
